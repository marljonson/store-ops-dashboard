```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load data
df = pd.read_csv('../data/store_data.csv', parse_dates=['week'])

# Calculate KPIs
df['sales_change'] = df.groupby('store_id')['sales'].pct_change()
df['labor_efficiency'] = df['sales'] / df['labor_hours']

# Summarize adoption impact
impact = df.groupby('initiative_adopted')[['sales', 'labor_efficiency']].mean().reset_index()

print(impact)

# Plot results
sns.barplot(x='initiative_adopted', y='sales', data=impact)
plt.title("Average Sales: Before vs After Initiative")
plt.show()

sns.barplot(x='initiative_adopted', y='labor_efficiency', data=impact)
plt.title("Labor Efficiency: Before vs After Initiative")
plt.show()
```
